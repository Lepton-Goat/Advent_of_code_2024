```python

import numpy as np 
import re as re

import time as t

t0 = t.time()

f = open('input.txt')
lines = []

for line in f:
    lines.append(line)


def search_vertical(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(0,len(array[0])):
            if array[i][j] == 'X':
                if array[i+1][j] == 'M':
                    if array[i+2][j] == 'A':
                        if array[i+3][j] == 'S':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def search_reverse_vertical(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(0,len(array[0])):
            if array[i][j] == 'S':
                if array[i+1][j] == 'A':
                    if array[i+2][j] == 'M':
                        if array[i+3][j] == 'X':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def search_diagonal(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(0,len(array[0])):
            if array[i][j] == 'X':
                if array[i+1][j+1] == 'M':
                    if array[i+2][j+2] == 'A':
                        if array[i+3][j+3] == 'S':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def search_reverse_diagonal(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(0,len(array[0])):
            if array[i][j] == 'S':
                if array[i+1][j+1] == 'A':
                    if array[i+2][j+2] == 'M':
                        if array[i+3][j+3] == 'X':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def search_back_diagonal(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(0,len(array[0])):
            if array[i][j] == 'X':
                if array[i+1][j-1] == 'M':
                    if array[i+2][j-2] == 'A':
                        if array[i+3][j-3] == 'S':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def search_reverse_back_diagonal(array):
    count = 0
    for i in range(0,len(array)-3):
        for j in range(3,len(array[0])):
            if array[i][j] == 'S':
                if array[i+1][j-1] == 'A':
                    if array[i+2][j-2] == 'M':
                        if array[i+3][j-3] == 'X':
                            count +=1
                        else:
                            continue
                    else:
                        continue
                else:
                    continue
            else:
                continue
    return count

def count_array(array):
    count = 0
    for i in range(0,len(array)):
        pattern = r'XMAS'
        horizontal = re.findall(pattern,array[i])
        count += len(horizontal)
    return count

def count_reverse(array):
    count = 0
    for i in range(0,len(array)):
        pattern = r'SAMX'
        horizontal = re.findall(pattern,array[i])
        count += len(horizontal)
    return count

vert = search_vertical(lines)

revert = search_reverse_vertical(lines)

diag = search_diagonal(lines)

revetdiag = search_reverse_diagonal(lines)

backdiag = search_back_diagonal(lines)

revertbackdiag = search_reverse_back_diagonal(lines)

horz = count_array(lines)

rev = count_reverse(lines)

# print(vert)
# print(revert)
# print(diag)
# print(revetdiag)
# print(backdiag)
# print(revertbackdiag)
# print(horz)
# print(rev)
print(vert + revert + diag + revetdiag + backdiag + revertbackdiag + horz + rev)

t1 = t.time()

print(t1-t0)
```
